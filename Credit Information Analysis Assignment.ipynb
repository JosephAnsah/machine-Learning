# [Problem 4] Data exploration

```python
%matplotlib inline
import seaborn as sns
# Make a heat map of the correlation coefficient matrix and find the relationships with high correlation coefficients.
corrmat = pd_applications.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
# Select features with high correlation with the target variable
number_of_selected_variables = 15
selected_cols = corrmat.nlargest(number_of_selected_variables,'TARGET')['TARGET'].index
correlation_coefficient = np.corrcoef(pd_applications[selected_cols].values.T)
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(correlation_coefficient, vmax=.8, square=True, yticklabels=selected_cols.values, xticklabels=selected_cols.values,annot=True)
```

```python
# check the correlation between the AMT_CREDIT and TARGET
AT_corr = pd_applications['AMT_CREDIT'].corr(pd_applications['TARGET'])
AT_corr
```

```python
# what type of loan paid mostly late
target_cash_1 = len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Cash loans'][pd_applications['TARGET']==1])
target_cash_2 = len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Cash loans'][pd_applications['TARGET']==0])

target_rev_1 = len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Revolving loans'][pd_applications['TARGET']==0])
target_rev_2 = len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Revolving loans'][pd_applications['TARGET']==1])

x = [1, 1.5]
y1, y2 = [target_cash_1, target_cash_2], [target_rev_1, target_rev_2]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

w = 0.2
ax.bar(x, y1, width=w, label='Cash')
ax.bar(np.array(x) + w, y2, width=w, label='Revolving')
ax.set_xticks([1.1, 1.6])
ax.set_xticklabels(['target0', 'target1'])
ax.legend()
plt.show()
```

```python
#checking with pie design

fig = plt.figure(figsize=(10,5))
plt.style.use('ggplot')
labels = ['Cash', 'Resolving']

sizes2 = [target_cash_1, target_rev_1]
ax2 = fig.add_subplot(1, 2, 1)
ax2.pie(sizes2, labels=labels, autopct='%1.0f%%')
ax2.set_title('TARGET=1')

sizes1 = [target_cash_2, target_rev_2]
ax1 = fig.add_subplot(1, 2, 2)
ax1.pie(sizes1, labels=labels, autopct='%1.0f%%')
ax1.set_title('TARGET=0')

plt.show()
```

```python
# check whether people who owns cars are mostly likely to delay payment of their loans
target_own_car_1 = len(pd_applications[pd_applications['FLAG_OWN_CAR']=='Y'][pd_applications['TARGET']==1])
target_own_car_2 = len(pd_applications[pd_applications['FLAG_OWN_CAR']=='Y'][pd_applications['TARGET']==0])

target_doesnt_own_car_1 = len(pd_applications[pd_applications['FLAG_OWN_CAR']=='N'][pd_applications['TARGET']==1])
target_doesnt_own_car_2 = len(pd_applications[pd_applications['FLAG_OWN_CAR']=='N'][pd_applications['TARGET']==0])

x = [1, 1.5]
y1, y2 = [target_own_car_1, target_own_car_2], [target_doesnt_own_car_1, target_doesnt_own_car_2]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

w = 0.2
ax.bar(x, y1, width=w, label='Own Car')
ax.bar(np.array(x) + w, y2, width=w, label="Deosn't own car")
ax.set_xticks([1.1, 1.6])
ax.set_xticklabels(['target0', 'target1'])
ax.legend()
plt.show()
```

```python
#checking with pie design

fig = plt.figure(figsize=(10,5))
plt.style.use('ggplot')
labels = ['Own Car', 'Doesn"t own car']

sizes2 = [target_own_car_1, target_doesnt_own_car_1]
ax2 = fig.add_subplot(1, 2, 1)
ax2.pie(sizes2, labels=labels, autopct='%1.0f%%')
ax2.set_title('TARGET=1')

sizes1 = [target_own_car_2, target_doesnt_own_car_2]
ax1 = fig.add_subplot(1, 2, 2)
ax1.pie(sizes1, labels=labels, autopct='%1.0f%%')
ax1.set_title('TARGET=0')

plt.show()
```
