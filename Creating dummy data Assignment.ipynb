# [Problem 1] Creating random numbers

```python
import numpy as np

mean = (-3, 0)
covariance = [[1.0, 0.8], [0.8,1.0]]

np.random.seed(0)

random_nbr = np.random.multivariate_normal(mean, covariance, 500)

print(random_nbr)
random_nbr.shape
```

# [Problem 2] Visualization by scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

mean = (-3, 0)
covariance = [[1.0, 0.8], [0.8,1.0]]

np.random.seed(0)

x, y = np.random.multivariate_normal(mean, covariance, 500).T

plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x, y)
plt.axis("equal")
plt.show()

```

# [Problem 3] Visualization by histogram

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Histogram of X1")
plt.xlabel("x1")
plt.ylabel("Frequency")

#mean = (-3, 0)
#covariance = [[1.0, 0.8], [0.8,1.0]]

x, y = np.random.multivariate_normal(mean, covariance, 500).T
plt.hist(x, color = "grey")
plt.xlim(right = 3)
plt.show()

plt.title("Histogram of X2")
plt.xlabel("x2")
plt.ylabel("Frequency")

x, y = np.random.multivariate_normal(mean, covariance, 500).T
plt.hist(x)
plt.xlim(left = -8)
plt.show()

```

# [Problem 4] Addition of data

```python
import matplotlib.pyplot as plt
import numpy as np

mean = (-3, 0)
covariance = [[1.0, 0.8], [0.8,1.0]]
mean2 = (0, -3)
covariance2 = [[1.0, 0.8], [0.8,1.0]]

np.random.seed(0)

x, y = np.random.multivariate_normal(mean, covariance, 500).T
x2, y2 = np.random.multivariate_normal(mean2, covariance2, 500).T

plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x, y, alpha = 0.5, label = "0")
plt.scatter(x2, y2, color = "orange", label = "1")
plt.legend(loc = "upper right")
plt.show()
```

# [Problem 5] Data combination

```python

random_nbr1 = np.random.multivariate_normal(mean, covariance, 500)
random_nbr2 = np.random.multivariate_normal(mean2, covariance2, 500)

np.random.seed(0)
random_nbrs_combined = np.vstack((random_nbr1, random_nbr2))

print(random_nbrs_combined)
random_nbrs_combined.shape
```

# [Problem 6] Labeling

```python
zeros_ndarray = np.zeros(500)
ones_ndarray = np.ones(500)

first_ndarray = np.insert(random_nbr1, 2, zeros_ndarray, axis = 1)
second_ndarray = np.insert(random_nbr2, 2, ones_ndarray, axis = 1)
np.random.seed(0)

lbl_ndarray = np.concatenate((first_ndarray, second_ndarray))

print(lbl_ndarray)
lbl_ndarray.shape
```
